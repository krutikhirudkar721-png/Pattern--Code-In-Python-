```python
# 15. Python program to reverse a given string (5 different ways).
s = "Python"

# 1. Slicing
print(s[::-1])

# 2. reversed()
print("".join(reversed(s)))

# 3. Loop
rev = ""
for ch in s:
    rev = ch + rev
print(rev)

# 4. While loop
i = len(s) - 1
while i >= 0:
    print(s[i], end="")
    i -= 1
print()

# 5. Recursion
def reverse(str):
    if len(str) == 0:
        return str
    return reverse(str[1:]) + str[0]

print(reverse(s))
```

```python
# 16. Python program to reverse a string using stack and reversed method.
s = "Python"

# Using stack
stack = list(s)
rev = ""
while stack:
    rev += stack.pop()
print(rev)

# Using reversed()
print("".join(reversed(s)))
```

```python
# 7. Count number of upper and lower case letters in string

def count_case(s):
    upper = 0
    lower = 0

    for i in s:
        if i.isupper():
            upper += 1
        elif i.islower():
            lower += 1

    print("No. of Upper case characters :", upper)
    print("No. of Lower case characters :", lower)

count_case("The quick Brow Fox")
```

```python
# 11. Write a Python function to check whether a number is Perfect or not

def perfect(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i

    if s == n:
        print("Perfect Number")
    else:
        print("Not Perfect Number")

perfect(6)
```

```python
# 12. Write a Python function to check whether a string is palindrome or not

def palindrome(s):
    if s == s[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome("madam")
```

```python
# 13. Write a Python function to print first n rows of Pascal's triangle

def pascal(n):
    for i in range(n):
        num = 1
        for j in range(i+1):
            print(num, end=" ")
            num = num * (i-j) // (j+1)
        print()

pascal(5)
```

```python
# 14. Write a Python function to check whether a string is pangram or not

import string

def pangram(s):
    alphabet = set(string.ascii_lowercase)
    if alphabet <= set(s.lower()):
        print("Pangram")
    else:
        print("Not Pangram")

pangram("The quick brown fox jumps over the lazy dog")
```
